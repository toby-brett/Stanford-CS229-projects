```python
import pandas as pd

file = pd.read_csv('/content/drive/My Drive/Ice Cream Sales and Temperature.csv')
X = file.iloc[:, 2]
y = file.iloc[:, 3]
print(y.head())
```

```python
import numpy as np

X_train = X.iloc[:40].to_numpy()
X_test = X.iloc[40:].to_numpy()
y_train = y.iloc[:40].to_numpy()
y_test = y.iloc[40:].to_numpy
print(X_train)
```

```python
# b(x) = 1 / x!
# eta = log lamda
# a(eta) = exp(eta)
#Â mean = lamba = exp(eta)
# can be found by differentiating A(eta) but A(eta) = exp(eta) its own derivative
```

```python
class GLM:
  def __init__(self):
    self.alpha = 0.00001
    self.theta = np.random.randn(1) * 0.01
  def forward(self, x):
    self.eta = self.theta * x # single len vec so dotting is not required
    self.mean = np.exp(self.eta)
    return self.mean
  def update(self, x, y):
    self.pred = self.forward(x)
    self.theta += self.alpha * (y - self.pred) * x

model = GLM()
```

```python
for i in range(100):
  for X, y in zip(X_train, y_train):
    model.update(X, y)
    print('prediction:', round(model.forward(X).item()), 'Real:', y, 'temp:', X, 'difference:', abs(y-round(model.forward(X).item())))
```
